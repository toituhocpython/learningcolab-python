  ```python
  tuoi = 17
  ```

- Dòng lệnh `print("Tôi tên là:",Ten,"\nTuổi của tôi:",Tuoi)` sẽ:
    - In chữ “Tôi tên là:” rồi giá trị biến `Ten`.
    - Sau đó xuống dòng nhờ ký tự đặc biệt `\n`, tiếp đến là “Tuổi của tôi:” và biến `Tuoi`.


# Nền tảng về biến trong Python

## 1. Biến là gì?
- **Biến** là một "tên" đại diện cho vùng nhớ dùng để lưu trữ giá trị (số, chuỗi, kết quả phép tính…).
- Biến giúp bạn tái sử dụng, xử lý dữ liệu mà không cần gán lại giá trị nhiều lần.

## 2. Quy tắc đặt tên biến
- Tên biến chỉ gồm chữ cái (a–z, A–Z), số (0–9) và dấu gạch dưới (_).
- **Không bắt đầu bằng số.**
- **Không dùng khoảng trắng.**
- Phân biệt chữ hoa–thường: `So` khác `so`.
- Ví dụ tên biến hợp lệ: `tuoi`, `ten_hoc_sinh`, `_giatrimax`

## 3. Gán giá trị cho biến
- Dùng dấu `=` để gán:  
  ```python
  x = 10
  y = "Mai"
  tuoi_hs = 17
  ```
- Biến có thể thay đổi giá trị bất cứ lúc nào.
  ```python
  x = 5
  x = 9  # x giờ là 9
  ```

## 4. Kiểu dữ liệu của biến
- Python tự động nhận biết kiểu dữ liệu khi gán giá trị: số nguyên, số thực, chuỗi, boolean.
- Kiểm tra kiểu dữ liệu bằng hàm `type()`:
  ```python
  x = 8.5
  print(type(x))   # Kết quả: <class 'float'>
  ```

## 5. Gán nhiều biến một lúc
```python
a, b, c = 1, 2, 3
print(a, b, c)  # Kết quả: 1 2 3
```

## 6. Đổi giá trị giữa hai biến
- Dùng phép hoán đổi dễ dàng:
  ```python
  a = 5
  b = 7
  a, b = b, a
  print(a, b)   # Kết quả: 7 5
  ```

## 7. Nhập giá trị từ người dùng
- Dùng hàm `input()` để lấy dữ liệu nhập vào:
  ```python
  ten = input("Nhập tên của bạn: ")
  ```
- Giá trị nhập vào luôn là kiểu chuỗi (str).

## 8. Ghi nhớ biến – tóm tắt dễ nhớ
- Biến = tên (tự đặt) + giá trị (nội dung muốn lưu).
- Tất cả thao tác gán/đặt lại giá trị đều làm thay đổi dữ liệu bên trong biến đó.
- Quy tắc đặt tên: Không có số ở đầu, không được có dấu cách, nên rõ ràng và dễ đọc.

## Tổng hợp các hàm đã học

Dưới đây là tổng hợp chi tiết **tên hàm**, **cách dùng**, và **lưu ý quan trọng** cho từng hàm Python mà bạn đã học trong buổi học này:

### 1. Hàm `print()`

**Cách dùng:**
```python
print(gia_tri_1, gia_tri_2, ..., sep=" ", end="\n")
```
- `gia_tri_1, gia_tri_2, ...`: Các biến hoặc giá trị cần in ra. Có thể in nhiều đối tượng cùng lúc, ngăn cách bằng `sep`.
- `sep`: (tuỳ chọn) đặt ký tự phân cách giữa các giá trị (mặc định là dấu cách)
- `end`: (tuỳ chọn) đặt ký tự kết thúc mỗi lần in (mặc định là xuống dòng `\n`)

**Ví dụ:**
```python
print("Xin chào!")
print("A", "B", sep="--")     # Xuất: A--B
print("Cột 1\tCột 2")           # Xuất: Cột 1    Cột 2 (tab)
print("Dòng 1\nDòng 2")         # Xuống dòng
```

**Lưu ý:**
- Có thể in nhiều giá trị cùng lúc, tự động thêm dấu cách nếu không đặt `sep` khác.
- Tham số `end` thay đổi ký tự cuối hàng (ví dụ, print không xuống dòng liên tục).
- Có thể truyền các biến khác kiểu trong một lệnh print, Python sẽ tự chuyển đổi sang chuỗi khi in.
- Đừng quên: Ký tự đặc biệt như `\n` xuống dòng, `\t` tab.

### 2. Hàm `input()`

**Cách dùng:**
```python
bien = input("Lời nhắc để người dùng nhập: ")
```
- Khi chạy, chương trình sẽ tạm dừng và chờ người dùng nhập dữ liệu, nhấn Enter để tiếp tục. Giá trị nhập vào luôn là **chuỗi** (str).[1][2][3]

**Ví dụ:**
```python
ten = input("Nhập tên của bạn: ")
print("Xin chào", ten)
```

**Lưu ý:**
- Luôn trả về **chuỗi**! Nếu cần kiểu số thì bạn phải "ép kiểu".
- Không truyền tham số theo dạng từ khóa (`prompt=`), chỉ truyền trực tiếp chuỗi lời nhắc (`input('...')`), nếu không sẽ lỗi cú pháp.[5]

### 3. Hàm ép kiểu: `int()`, `float()`, `str()`

**a) `int()`** — chuyển đổi giá trị sang kiểu **số nguyên**
```python
n = int("8")       # n = 8, kiểu int
```
- Nếu truyền giá trị không hợp lệ (ví dụ "abc") sẽ xảy ra lỗi.

**b) `float()`** — chuyển đổi giá trị sang kiểu **số thực**
```python
f = float("3.14")  # f = 3.14, kiểu float
```

**c) `str()`** — chuyển giá trị bất kỳ sang kiểu **chuỗi**
```python
s = str(9.2)        # s = "9.2"
```

**Lưu ý chung:**
- Khi nhận dữ liệu từ `input()` mà cần dùng kiểu số, luôn ép kiểu lại: `tuoi = int(input("Nhập tuổi: "))`
- Nếu giá trị truyền vào không chuyển đổi được sang số, sẽ ra lỗi (ValueError). Hãy kiểm soát dữ liệu đầu vào khi làm bài lớn hoặc thực tế.

***

### 4. Hàm kiểm tra kiểu dữ liệu: `type()`

**Cách dùng:**
```python
print(type(bien))
```
- Trả về loại kiểu dữ liệu của biến (`<class 'int'>`, `<class 'float'>`, `<class 'str'>`, ...).

**Lưu ý:**
- Chỉ nhận **một biến** mỗi lần gọi hàm.
- Rất hữu ích để kiểm tra kiểu trong quá trình học, thử nghiệm các loại dữ liệu.

***

### Tóm tắt rất ngắn gọn
- **print()**: In ra màn hình, hỗ trợ nhiều giá trị, tuỳ biến `sep` và `end`.
- **input()**: Nhận vào dữ liệu từ bàn phím, luôn là chuỗi, dùng trong chương trình tương tác.[2][3][1][5]
- **int()/float()/str()**: Ép kiểu, chuyển đổi dữ liệu thường dùng sau khi nhận input.[3][2]
- **type()**: Kiểm tra kiểu dữ liệu của một biến.

Khi cần bài tập hoặc ví dụ mở rộng cho từng hàm, chỉ cần nói nhé!



````python

#Bài tập về kiểu dữ liệu
songuyen = 1
sothuc = 3.14
chuoi = "python ngay 1"
booldung = True
boolsai = False
print(
    "\nclass cua songuyen la:",type(songuyen),
    "\nclass cua so thuc la:",type(sothuc),
    "\nclass cua chuoi la:",type(chuoi),
    "\nclass cua booldung la:",type(booldung),
    "\nclass cua boolsai la:",type(boolsai)
)

ten = input("Nhập tên của bạn: ")
age_str = input("Nhập tuổi của bạn: ")
age = int(age_str)
print(
    "Tên của bạn là: ", ten,
    "\nTuổi của bạn là: ",age
)
````
